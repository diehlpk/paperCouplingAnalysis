```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
pgf_with_latex = {"text.usetex": True, "font.size" : 12, "pgf.preamble" : [r'\usepackage{xfrac}'] }
```

```python
x = np.linspace(0,3,100)
```

```python
plt.plot(x,np.sin(x*np.pi*2),c="black")
plt.grid()
plt.xlabel(r"$x$")
plt.ylabel(r"$u(x)$")
plt.savefig("u-sin.pdf",bbox_inches='tight')
```

```python
plt.plot(x,np.cos(x*np.pi*2+np.pi/2),c="black")
plt.grid()
plt.xlabel(r"$x$")
plt.ylabel(r"$u(x)$")
plt.savefig("u-cos.pdf",bbox_inches='tight')
```

```python
plt.plot(x,np.tan(x),c="black")
plt.grid()
plt.xlabel(r"$x$")
plt.ylabel(r"$u(x)$")
plt.savefig("u-tan.pdf",bbox_inches='tight')
```
